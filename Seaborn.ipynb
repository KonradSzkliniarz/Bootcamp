```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

# TIPS DATA

```
df= sns.load_dataset('tips')
df.head()
```

```
df.info()
```

```
df.describe().T
```

```
df.describe(include=['category']).T
```

# Relacja wielkości napiwku do rachunku

```
sns.relplot(data=df,x='total_bill',y='tip',height=7)
```

### zmiana wielkości czcionki

```
sns.set(font_scale=1.2)
sns.relplot(data=df,x='total_bill',y='tip',height=7)
```

```
sns.relplot(data=df,x='total_bill',y='tip',size='size',height=7, hue='size')
```

```
sns.relplot(data=df,x='total_bill',y='tip',size='size',height=7, hue='size', col='time')
```

```
sns.relplot(data=df,x='total_bill',y='tip',size='size',height=7, hue='time')
```

```
sns.relplot(data=df,x='total_bill',y='tip',size='size',height=7, hue='size', col='time', row='smoker')
```

```
sns.catplot(data=df,x='day',y='total_bill', height=8)
```

```
sns.catplot(data=df,x='day',y='total_bill', height=8, kind='swarm')
```

```
sns.catplot(data=df,x='day',y='total_bill', height=8, kind='box')
```

```
sns.catplot(data=df,x='day',y='total_bill', height=8, kind='violin')
```

```
sns.catplot(data=df,x='day',y='total_bill', height=8, kind='bar')
```

```
sns.catplot(data=df,x='day', height=8, kind='count')
```
