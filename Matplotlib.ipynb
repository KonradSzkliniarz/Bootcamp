# Tworzenie wielu wykresów

```
plt.subplot(121)
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2],[3,6,5],'r--')
```

```
fig = plt.figure(figsize=(13,9))

plt.subplot(131)
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2],[3,6,5],'r--')
plt.subplot(133)
plt.plot([0,1,2],[3,6,5],'ro')

fig.show()
```
